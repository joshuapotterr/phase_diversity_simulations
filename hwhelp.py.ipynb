```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
pi = np.pi
E1 = 1  # E1 is normalized
U0 = 4 * E1  # U0 in terms of E1

# Define the quantization condition
def quantization_condition(E):
    if E >= U0 or E <= 0:
        return np.inf  # Not valid for E outside this range
    return np.sqrt(E) * np.tan(np.sqrt(E / E1) * pi) + np.sqrt(U0 - E)

# Plot the quantization condition
E_values = np.linspace(0.1, U0 - 0.1, 1000)  # Avoid exact 0 and U0 to avoid infinities
quantization_values = [quantization_condition(E) for E in E_values]

plt.figure(figsize=(10, 6))
plt.plot(E_values / E1, quantization_values, label="Quantization Condition")
plt.axhline(0, color='red', linestyle='--', label="Zero Line")
plt.xlabel("Energy (E/E1)")
plt.ylabel("Quantization Condition")
plt.title("Quantization Condition vs Energy")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constants (SI units; you can adjust them if needed):
hbar = 1.054571817e-34  # Reduced Planck constant [J·s]
k_B  = 1.380649e-23     # Boltzmann constant [J/K]

# Example parameters:
omega = 1.0e13          # Angular frequency [rad/s] (adjust to suit your problem)
N = 1.0                 # Number of particles (adjust as needed)

# Temperature range:
T_min = 1.0      # minimum temperature in K
T_max = 500.0    # maximum temperature in K
num_points = 500

T_values = np.linspace(T_min, T_max, num_points)

def Cv(T):
    """Specific heat as a function of temperature T."""
    x = hbar * omega / (k_B * T)
    return N * k_B * x**2 * np.exp(-x)

Cv_values = Cv(T_values)

# Plot the results:
plt.figure(figsize=(7, 5))
plt.plot(T_values, Cv_values, label=r'$C_v(T)$', color='blue')
plt.xlabel('Temperature (K)')
plt.ylabel(r'$C_v$ (J/K)')
plt.title(r'Specific Heat $C_v = N k_B \left(\frac{\hbar \omega}{k_B T}\right)^2 e^{-\hbar \omega / (k_B T)}$')
plt.grid(True)
plt.legend()
plt.show()
```
